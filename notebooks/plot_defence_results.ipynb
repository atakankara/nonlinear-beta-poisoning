```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open("best_cifar_precision_hist.pckl", "rb") as f:
    cifar_precision_hist = pickle.load(f)

with open("best_cifar_recall_hist.pckl", "rb") as f:
    cifar_recall_hist = pickle.load(f)
    
with open("best_cifar_f1_hist.pckl", "rb") as f:
    cifar_f1_hist = pickle.load(f)
    
with open("best_cifar_accuracy_hist.pckl", "rb") as f:
    cifar_accuracy_hist = pickle.load(f)
```

```python
with open("best_mnist_precision_hist.pckl", "rb") as f:
    mnist_precision_hist = pickle.load(f)

with open("best_mnist_recall_hist.pckl", "rb") as f:
    mnist_recall_hist = pickle.load(f)
    
with open("best_mnist_f1_hist.pckl", "rb") as f:
    mnist_f1_hist = pickle.load(f)
    
with open("best_mnist_accuracy_hist.pckl", "rb") as f:
    mnist_accuracy_hist = pickle.load(f)
```

```python
threshold_list = np.arange(-5.01, 5.01, 0.01)

plt.rcParams.update({'font.size': 16})
x_ticks = np.arange(-5.01, 5.01, 2)
y_ticks = np.arange(0.0, 1.01, 0.2)
font_size = 22
line_width = 2.5
cifar_legend = "CIFAR-10"
mnist_legend = "MNIST"
figure, axis = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

axis[0,0].plot(threshold_list, mnist_precision_hist, c="b", linewidth=2.5, label=mnist_legend)
axis[0,0].plot(threshold_list, cifar_precision_hist, c="r", linewidth=2.5, label=cifar_legend, linestyle="--")
axis[0,0].xaxis.set_ticks(x_ticks)
axis[0,0].yaxis.set_ticks(y_ticks)
axis[0,0].set_xlabel(r"$\tau$", fontsize=font_size)
axis[0,0].set_ylabel("Precision", fontsize=font_size)
axis[0,0].legend()

axis[0,1].plot(threshold_list, mnist_recall_hist, c="b", linewidth=2.5, label=mnist_legend)
axis[0,1].plot(threshold_list, cifar_recall_hist, c="r", linewidth=2.5, label=cifar_legend, linestyle="--")
axis[0,1].xaxis.set_ticks(x_ticks)
axis[0,1].yaxis.set_ticks(y_ticks)
axis[0,1].set_xlabel(r"$\tau$", fontsize=font_size)
axis[0,1].set_ylabel("Recall", fontsize=font_size)
axis[0,1].legend()

axis[1,0].plot(threshold_list, mnist_f1_hist, c="b", linewidth=2.5, label=mnist_legend)
axis[1,0].plot(threshold_list, cifar_f1_hist, c="r", linewidth=2.5, label=cifar_legend, linestyle="--")
axis[1,0].xaxis.set_ticks(x_ticks)
axis[1,0].yaxis.set_ticks(y_ticks)
axis[1,0].set_xlabel(r"$\tau$", fontsize=font_size)
axis[1,0].set_ylabel("F1-Score", fontsize=font_size)
axis[1,0].legend()

axis[1,1].plot(threshold_list, mnist_accuracy_hist, c="b", linewidth=2.5, label=mnist_legend)
axis[1,1].plot(threshold_list, cifar_accuracy_hist, c="r", linewidth=2.5, label=cifar_legend, linestyle="--")
axis[1,1].xaxis.set_ticks(x_ticks)
axis[1,1].yaxis.set_ticks(y_ticks)
axis[1,1].set_xlabel(r"$\tau$", fontsize=font_size)
axis[1,1].set_ylabel("Accuracy", fontsize=font_size)
axis[1,1].legend()

figure.suptitle(r"Results of Discriminator-based defence with different $\tau$ parameters")
figure.tight_layout()
plt.savefig("tau_results.jpg")
```
