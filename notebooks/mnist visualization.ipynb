```python
from sklearn.datasets import fetch_openml
from matplotlib import pyplot as plt
from sklearn import decomposition
from sklearn.manifold import TSNE
import numpy as np
import keras
```

```python
raw_data, raw_labels = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
nsamples = 5000
data = raw_data[:nsamples]
labels = raw_labels[:nsamples]
```

```python
data = data / 255.
labels = labels.astype('int')
images = data.to_numpy().reshape(data.shape[0], 28, 28)
```

```python
plt.figure(figsize=(8,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title('truth: {}'.format(labels[i]))
```

```python
pca = decomposition.PCA(n_components=2)
view = pca.fit_transform(data)
plt.scatter(view[:,0], view[:,1], c=labels, alpha=0.2, cmap='Set1')
```

```python
np.logical_or((labels == 4), (labels == 0))
```

```python
data[np.logical_or((labels == 4), (labels == 0))]
```

```python
view = TSNE(n_components=2, random_state=123).fit_transform(data)

plt.figure(figsize=(20,10))
plt.scatter(view[:,0], view[:,1], c=labels, alpha=0.5)
plt.title("MNIST on 10 digits")
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')
```

```python
view = TSNE(n_components=2, random_state=123).fit_transform(data[np.logical_or((labels == 4), (labels == 0))])

plt.figure(figsize=(20,10))
plt.scatter(view[:,0], view[:,1], c=labels[np.logical_or((labels == 4), (labels == 0))], alpha=0.5)
plt.title("MNIST on 4-0")
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')
```
