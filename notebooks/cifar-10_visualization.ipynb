```python
from sklearn.datasets import fetch_openml
from matplotlib import pyplot as plt
from sklearn import decomposition
from sklearn.manifold import TSNE
import numpy as np
import tensorflow as tf
from torchvision import datasets,transforms
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
nsamples = 5000
data = x_train[:nsamples]
labels = y_train[:nsamples]
```

```python
data = data / 255.
data = data.reshape(data.shape[0], -1)
labels = labels.astype('int')
images = data
```

```python
pca = decomposition.PCA(n_components=2)
view = pca.fit_transform(data)
plt.scatter(view[:,0], view[:,1], c=labels, alpha=0.2, cmap='Set1')
```

```python
view = TSNE(n_components=2, random_state=123).fit_transform(data)

plt.figure(figsize=(20,10))
plt.scatter(view[:,0], view[:,1], c=labels, alpha=0.5)
plt.title("CIFAR on 10 classes")
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')
```

```python
view = TSNE(n_components=2, random_state=123).fit_transform(data[np.logical_or((labels.flatten() == 6), (labels.flatten() == 8))])

plt.figure(figsize=(20,10))
plt.scatter(view[:,0], view[:,1], c=labels[np.logical_or((labels == 6), (labels == 8))], alpha=0.5)
plt.title("CIFAR-10 on frog-ship")
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')
```
